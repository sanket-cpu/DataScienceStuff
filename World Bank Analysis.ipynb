# Data Cleaning and Preprocessing 


```python
df.shape
df.isna()
```

```python
peekdata = df.iloc[: , 3:].copy()
```

```python
info_missing_values= peekdata.isna().sum() / len(peekdata)
```

```python
print(info_missing_values)
```

```python
type(info_missing_values) # as we can see - we get a Pandas series as an output 
```

```python
info_missing_values.dtype # we know that the series is of a float type 
```

```python
s= info_missing_values [ info_missing_values >= 0.5] 
#filtering the columns that have more than 80 percent NaN values
```

```python
s
```

```python
#since we have peekdata as our dataset i will drop the above columns in peekdata

df = df.drop(columns = s.keys())
```
